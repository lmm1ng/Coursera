```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
a = 1 # Параметр для распределения
b = 4 # Параметр для распределения
beta_rv = sts.beta(a,b) # Бета распределение
sample = beta_rv.rvs(1000) # Создаём выборку 1000
```

```python
x = np.linspace(0,1,1000) # Массив x-ов
plt.hist(sample, normed = True) # Строим гистограмму
pdf = beta_rv.pdf(x) # Плотность распределения
plt.plot(x,pdf) # График плотности
# Форматирование графика
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```

```python
# Определяем функцию для нормального распредения
def norm(n):
    print 'n = ' + str(n)
    a = 1.0
    b = 4.0
    sample = [np.mean(beta_rv.rvs(n)) for x in range(1000)] # 1000 Выборок объёма n
    plt.hist(sample, normed = True)
    math_ozh = a / (a + b) # Мат. ожидание бета распределения
    disp = np.sqrt(a*b/((a + b)**2 *(a + b + 1))/n) # Дисперсия
    norm_rv = sts.norm(loc = math_ozh, scale = disp) # Нормальное распределние относительно наших параметров
    x = np.linspace(0,1,1000)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
    plt.legend()
    plt.xlabel('$x$')
    plt.ylabel('$f(x)$')
    plt.show()
```

```python
norm(5) # 1000 объёма 5
```

```python
norm(10) # 1000 объёма 10
```

```python
norm(50) # 1000 выборок объёма 50
```

Распределение выборочных средних хорошо описывается нормальным распределением. С ростом объёма n точность аппроксимации увеличивается
