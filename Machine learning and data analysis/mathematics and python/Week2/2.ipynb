```python
import numpy as np
import codecs
from io import open
import scipy.linalg as lg
from matplotlib import pylab as plt
```

```python
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)
```

```python
def to_matrix_solve(mas):
    a =[]
    b =[]
    for point in mas:
        a.append([point**step for step in range(len(mas))])
        b.append(f(point))
    return a,b
```

```python
def solve_sys(matr):
    (a,b) = to_matrix_solve(matr)
    return lg.solve(a,b)
```

```python
def poly(mas, x):
    new_mas = solve_sys(mas)
    phi = np.zeros(len(x))
    for step in range(len(new_mas)):
        phi += new_mas[step] * x ** step
    return phi
```

```python
x = np.arange(1,16,0.01)
print x
y = f(x)
plt.plot(x,y)
plt.show()
```

```python
y = poly([1,15], x)
plt.plot(x,y)
plt.show()
```

```python
y = poly([1,8,15], x)
plt.plot(x,y)
plt.show()
```
